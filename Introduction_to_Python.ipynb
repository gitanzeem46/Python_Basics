```python
class vehical:
    def __init__(self,mileage,cost):
        self.mileage=mileage
        self.cost=cost
    def show_vehical_details(self):
        print("i am vehical")
        print("mileage of vehical is ",self.mileage)
        print("cost of vehical is ",self.cost)
        print("i am vehical")
```

```python
v1=vehical(300,800000)
```

```python
class car(vehical):
    def __init__(self,mileage,cost,tyres,hp):
        super().__init__(mileage,cost)
        self.tyres=tyres
        self.hp=hp
    def show_car_details(self):
        print("i am a car")
        print("number of tyres are",self.tyres)
        print("horse power is ",self.hp)
```

```python
# multiple inheritance
```

```python
class parent1():
    def assign_string_one(self,str1):
        self.str1=str1
    def show_string_one(self):
        return self.str1
```

```python
class parent2():
    def assign_string_two(self,str2):
        self.str2=str2
    def show_string_two(self):
        return self.str2
```

```python
class child (parent1,parent2):
    def assign_string_three(self,str3):
        self.str3=str3
    def show_string_three(self):
        return self.str3
```

```python
my_child=child()
```

```python
my_child.assign_string_one("i am string of parent 1")
```

```python
my_child.assign_string_two("i am string of parent 2")
```

```python
my_child.assign_string_three("i am string of parent 3")
```

```python
my_child.show_string_one()
```

```python
my_child.show_string_two()
```

```python
my_child.show_string_three()
```

```python
# multi-level inheritance
```

```python
class parent():
    def get_name(self,name):
        self.name=name
    def show_name(self):
        return self.name
```

```python
class child(parent):
    def get_age(self,age):
        self.age=age
    def show_age(self):
        return self.age
    
```

```python
class grandchild(child):
    def get_gender(self,gender):
        self.gender=gender
    def show_gender(self):
        return self.gender
```

```python
gc=grandchild()
```

```python
gc.get_name('Aman')
```

```python
gc.get_age(21)
```

```python
gc.get_gender('Male')
```

```python
gc.show_name()
```

```python
gc.show_age()
```

```python
gc.show_gender()
```
